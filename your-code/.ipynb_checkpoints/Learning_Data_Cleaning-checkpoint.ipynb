```python
DataFrameInicial[['displ', 'cylinders']] = DataFrameInicial[['displ', 'cylinders']].fillna(0)
#.fillna(0) llena los NA por 0 (le puedes poner entre parentesis el valor que quieras)

DataFrameInicial
```

```python
#Verificar si hay valores nulos espúreos
MuestraPrueba = DataFrameInicial[(DataFrameInicial['cylinders'] == 0) & (DataFrameInicial['displ']!=0)]
MuestraPrueba     #Te muestra los cilindros que son cero y que no tienen un display igual a cero. Entonces es espúreo porque #cars
```

```python
DataFrameInicial.loc[(DataFrameInicial['cylinders']==0) & (DataFrameInicial['displ']!=0), 'cylinders'] = 4
#Lo que cumpla con los dos condicionees, "cylinders sean cero" y "display no sea 0", en la colulmna cylinders la vas a poner en 4

#Comprobación
dato = DataFrameInicial.loc[21506]
dato['cylinders']
```

```python
#Datos que varían poco (variabilidad baja)
import numpy as numpy

VariabilidadBaja = []

for col in DataFrameInicial._get_numeric_data():
    minimo = min(DataFrameInicial[col])
    percentil90 = numpy.percentile(DataFrameInicial[col],90)  #si 90% de los datos están agrupados en el valor minimo, entonces no varía y los quitamos
    if percentil90 == minimo:                                  #percentile es una función de numpy
        VariabilidadBaja.append(col)

print(VariabilidadBaja)
```

```python
DataFrameInicial = DataFrameInicial.drop(VariabilidadBaja, axis = 1)
#Limpia a los variables que no tienen variabiliad
```

```python
#Para describir tu base datos
DataFrameInicial.describe()
```

```python
Analisis = DataFrameInicial.describe().transpose()  #Si quieres cambiar la presentación de datos
Analisis
```

```python
#Crear columna para hacer calcula entre dif percentil 75 y 25
Analisis['Intercuartil'] = Analisis['75%'] - Analisis['25%']
Analisis
```

```python
Resultado = pd.DataFrame(columns=DataFrameInicial.columns)
Resultado
```

```python
for row in Analisis.index: 
    intercuartil = Analisis.at[row, "Intercuartil"]  #.at significa "en" y busca en row y en la columna "Intercuartil"
    temporal = intercuartil * 1.5                
    baja = Analisis.at[row, "25%"] - temporal   
    alta = Analisis.at[row, "75%"] + temporal
    resultados = DataFrameInicial[(DataFrameInicial[row]<baja) | (DataFrameInicial[row]>alta)].copy() 
    resultados['Resultados'] = row
    Resultado = Resultado.append(resultados)

Resultado
```

```python
#Reemplazando valores para limpiarlos
DataFrameInicial['trany'] = DataFrameInicial['trany'].str.replace('-', ' ')
DataFrameInicial['trany'] = DataFrameInicial['trany'].str.replace('Automatic', 'Auto')
DataFrameInicial['trany'] = DataFrameInicial['trany'].str.replace('Auto ', 'Auto')
DataFrameInicial['trany'] = DataFrameInicial['trany'].str.replace(' ', '')
DataFrameInicial['trany'] = DataFrameInicial['trany'].str.replace('(', '')
DataFrameInicial['trany'] = DataFrameInicial['trany'].str.replace(')', '')


set(DataFrameInicial['trany']) #con esto quitaste guiones y homogenizaste los automatic a auto
```
